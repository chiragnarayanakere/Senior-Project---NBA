# 2011 draft class


```python
data1516 = pd.read_csv('Downloads/NBA Player Stats - Regular Stats 2015-2016.csv')
data1516.drop(['Tm', 'Rk'], axis=1, inplace=True)
data1415 = pd.read_csv('Downloads/NBA Player Stats - Regular Stats 2014-2015.csv')
data1415.drop(['Tm', 'Rk'], axis=1, inplace=True)
data1314 = pd.read_csv('Downloads/NBA Player Stats - Regular Stats 2013-2014.csv')
data1314.drop(['Tm', 'Rk'], axis=1, inplace=True)
data1213 = pd.read_csv('Downloads/NBA Player Stats - Regular Stats 2012-2013.csv')
data1213.drop(['Tm', 'Rk'], axis=1, inplace=True)
data1112 = pd.read_csv('Downloads/NBA Player Stats - Regular Stats 2011-2012.csv')
data1112.drop(['Tm', 'Rk'], axis=1, inplace=True)
data1516
```

# College data

```python
draft2011 = pd.read_csv('Downloads/part 2 first model - 2011 draft class.csv')
draft2011['Unnamed: 0'] = draft2011['Unnamed: 0'].str.split('\\')
for i in range(1, len(draft2011)):
    
    draft2011['Unnamed: 0'][i] = draft2011['Unnamed: 0'][i][0]

draft2011
```

# Avg. of 2010-11 to 2014-15

```python
names = list(draft2011['Unnamed: 0'][1:])
namenyear = draft2011[['Unnamed: 0', 'Unnamed: 1']][1:] 
namenyear = namenyear.sort_values('Unnamed: 0')
namenyear = namenyear.reset_index()
namenyear = namenyear.drop('index', 1)

#------------------------------------------------------------

n1516 = []
for i in range(len(data1516['Player'])):
    for j in range(len(names)):
        if data1516['Player'][i] == names[j]:
            n1516.append(names[j])       
n1516 = list(set(n1516))

f1516 = pd.DataFrame(index=range(len(n1516)),columns=data1516.columns)

for i in range(len(n1516)):

    for word in f1516.columns:

         f1516[word][i] = data1516.loc[data1516['Player'] == n1516[i]][word].values[0]

#------------------------------------------------------------

n1415 = []
for i in range(len(data1415['Player'])):
    for j in range(len(names)):
        if data1415['Player'][i] == names[j]:
            n1415.append(names[j])       
n1415 = list(set(n1415))

f1415 = pd.DataFrame(index=range(len(n1415)),columns=data1415.columns)

for i in range(len(n1415)):

    for word in f1415.columns:

         f1415[word][i] = data1415.loc[data1415['Player'] == n1415[i]][word].values[0]
                       
#------------------------------------------------------------

n1314 = []
for i in range(len(data1314['Player'])):
    for j in range(len(names)):
        if data1314['Player'][i] == names[j]:
            n1314.append(names[j])       
n1314 = list(set(n1314)) 

f1314 = pd.DataFrame(index=range(len(n1314)),columns=data1314.columns)

for i in range(len(n1314)):

    for word in f1314.columns:

         f1314[word][i] = data1314.loc[data1314['Player'] == n1314[i]][word].values[0]            
        
#------------------------------------------------------------

n1213 = []
for i in range(len(data1213['Player'])):
    for j in range(len(names)):
        if data1213['Player'][i] == names[j]:
            n1213.append(names[j])       
n1213 = list(set(n1213)) 

f1213 = pd.DataFrame(index=range(len(n1213)),columns=data1213.columns)

for i in range(len(n1213)):

    for word in f1213.columns:

         f1213[word][i] = data1213.loc[data1213['Player'] == n1213[i]][word].values[0] 
            
#------------------------------------------------------------

n1112 = []
for i in range(len(data1112['Player'])):
    for j in range(len(names)):
        if data1112['Player'][i] == names[j]:
            n1112.append(names[j])       
n1112 = list(set(n1112)) 

f1112 = pd.DataFrame(index=range(len(n1112)),columns=data1112.columns)

for i in range(len(n1112)):

    for word in f1112.columns:

         f1112[word][i] = data1112.loc[data1112['Player'] == n1112[i]][word].values[0] 

#------------------------------------------------------------

final = f1516.append([f1415, f1314, f1213, f1112])
final = final.reset_index()
final.drop(['index'], axis=1, inplace=True)

final['Age']= final['Age'].astype('float64')
final['G']= final['G'].astype('float64')
final['GS']= final['GS'].astype('float64')
final['MP']= final['MP'].astype('float64')
final['FG']= final['FG'].astype('float64')
final['FGA']= final['FGA'].astype('float64')
final['FG%']= final['FG%'].astype('float64')
final['3P']= final['3P'].astype('float64')
final['3PA']= final['3PA'].astype('float64')
final['3P%']= final['3P%'].astype('float64')
final['2P']= final['2P'].astype('float64')
final['2PA']= final['2PA'].astype('float64')
final['2P%']= final['2P%'].astype('float64')
final['eFG%']= final['eFG%'].astype('float64')
final['FT']= final['FT'].astype('float64')
final['FTA']= final['FTA'].astype('float64')
final['FT%']= final['FT%'].astype('float64')
final['ORB']= final['ORB'].astype('float64')
final['DRB']= final['DRB'].astype('float64')
final['TRB']= final['TRB'].astype('float64')
final['AST']= final['AST'].astype('float64')
final['STL']= final['STL'].astype('float64')
final['BLK']= final['BLK'].astype('float64')
final['TOV']= final['TOV'].astype('float64')
final['PF']= final['PF'].astype('float64')
final['PS/G']= final['PS/G'].astype('float64')

finaldata = final.groupby('Player', as_index=False).mean()
finaldata.insert(1, "Years in College", namenyear['Unnamed: 1']) 
finaldata['Years in College'] = 0

finaldata = finaldata[['Player', 'Years in College', 'Age', 'G', 'GS', 'MP', 'FG', 'FGA',
       'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA',
       'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PS/G']]
```

```python
x = []
for i in range(len(finaldata['Player'])):
    for j in range(len(n1516)):
        if finaldata['Player'][i] == n1516[j]:
            x.append(n1516[j])       
f = list(set(x))

test = finaldata

test = test[test['Player'].isin(f)].reset_index().drop('index', 1)

test.to_csv('Desktop/finaldata/2011draft.csv', sep='\t', encoding='utf-8')
test
```
