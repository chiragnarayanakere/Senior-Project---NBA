```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.plotly as py
import warnings
import xgboost as xgb
warnings.filterwarnings('ignore')
```

## 2017-2018 data

```python
data = pd.read_csv('Desktop/senior_project/2017-2018nba.csv')
data1718 = data.drop(['Rk'], axis=1)

#Removing duplicates, for players that got traded I'm going to get the average of all their stats so we can
#get a more comprehensive picture
data1718_final = data1718.groupby('Player', as_index=False).mean()
data1718_final.head()
```

```python
target1718 = data1718_final['PTS']
newdata1718 = data1718_final.drop(['PTS', 'Player'], axis=1)
```

```python
model1718 = xgb.XGBClassifier()
model1718.fit(newdata1718, target1718)
xgb.plot_importance(model1718, importance_type = 'weight')
plt.show()
```
